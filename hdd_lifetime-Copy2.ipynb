```python
# 최종일 별 데이터 수 count
# 제일 많은 건 2016-04-29 지만 숫자 차이가 너무 커서 그래프에선 제외

count_date = pd.value_counts(df_date['date'].values, sort=True)
count_date[1:].plot.bar()
count_date[1:].plot.bar(grid=True, figsize=(10,8), fontsize=15)  # figsize, fontsize 조정
```

```python
# 1월1일 데이터만 남기고 중복을 제거하자

df_startdate = df_train[df_train['date']=='2016-01-01']
df_startdate = df_startdate.drop_duplicates(['serial_number'])
df_startdate
```

```python
normalized_columns = [col for col in df_startdate.columns if 'normal' in col]
normalized_columns
```

```python
df_normalized = df_startdate[normalized_columns]
df_normalized
```

```python
# 결측치가 있는 컬럼 제거

df_normalized = df_normalized.dropna(axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df_scaled = sc.fit_transform(df_normalized)
df_scaled
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit(df_scaled)

df_normalized['cluster'] = clusters.labels_
df_normalized.head()
```

```python
df_startdate['cluster'] = clusters.labels_
df_startdate
```

```python
df_startdate.groupby('cluster').count()
```

```python
df_startdate.groupby('cluster').mean()
```

```python
from sklearn.decomposition import PCA
X = df_normalized

pca = PCA(n_components=2)

pca.fit(X)
x_pca = pca.transform(X)
x_pca
```

```python
pca_df = pd.DataFrame(x_pca)
pca_df['cluster'] = df_normalized['cluster']
pca_df.head()
```

```python
import seaborn as sns

for i in sorted(pca_df['cluster'].unique()):
    tmp = pca_df.loc[pca_df['cluster'] == i]
    plt.scatter(tmp[0], tmp[1])
    plt.legend(sorted(pca_df['cluster'].unique()))
```

```python
df_lifetime = pd.merge(df_startdate, df_date)
```

```python
df_lifetime
```
